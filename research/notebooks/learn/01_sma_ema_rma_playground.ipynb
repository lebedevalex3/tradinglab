```python
import pandas as pd

pd.set_option("display.width", 140)
pd.set_option("display.max_columns", 50)

# простой ряд: "ступенька", чтобы видеть отклик
x = pd.Series([0] * 20 + [10] * 20, name="x")
L = 10
```

```python
sma = x.rolling(L).mean()

# EMA: стандартная эксп. средняя (alpha = 2/(L+1))
ema = x.ewm(span=L, adjust=False).mean()

df = pd.DataFrame({"x": x, "sma": sma, "ema": ema})
df.head(25)
```
